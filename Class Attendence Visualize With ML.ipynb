```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
```

```python
# Load your data
df = pd.read_csv('Batch_Attendance.csv')
df.head()
```

```python
# find missing values in data
missing_values = df.isnull().sum()
missing_values
```

```python
# missing values in the "Type" column replace with "STUDENT"
df['Type'].fillna('STUDENT', inplace=True)
```

```python
# columns to replace missing values
columns_to_replace = ['07-01-21', '06-01-21', '05-01-21', '04-01-21', '03-01-21', '02-01-21', '01-01-21']

# Fill missing values in columns with "Attended"
for column in columns_to_replace:
    df[column].fillna('Attended', inplace=True)
```

```python
# find missing values in data after missing values replace
missing_values = df.isnull().sum()
missing_values
```

```python
# Visualize the 'Type' column (target variable)
plt.figure(figsize=(4, 2))
sns.countplot(data=df, x='Type')
plt.title('Distribution of Status (Target Variable)')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Extract the columns with dates 
df_columns = [col for col in df.columns if col[0].isdigit()]

# Transpose the data dates as rows 
df_total = df[df_columns].T

# Count the number of students who attended and missed on each date
attendance_counts = df_total.apply(lambda x: x.value_counts().get('Attended', 0))
missing_counts = df_total.apply(lambda x: x.value_counts().get('Missed', 0))

# Adding columns Attendence and Average Rating
df["Attended_count"] = attendance_counts
df["Missed_count"] = missing_counts
df['Average_Rating'] = df.filter(like='R-').mean(axis=1)
df.head(10)
```

# Students need attention for missing sessions

```python
# Identify students needing attention
students_needing_attention = df[df['Missed_count'] > 6]['Student Roll Num']
print("\nVery Satisfied Students: \n", students_needing_attention)
```

```python
# Create a bar chart for students needing attention
print("\nVery Satisfied Students: \n")
plt.figure(figsize=(10, 6))
sns.barplot(x=students_needing_attention, y=df[df['Student Roll Num'].isin(students_needing_attention)]['Missed_count'])
plt.xlabel('Students')
plt.ylabel('Sessions Missed')
plt.title('Students Needing Attention (Missed Sessions > 6)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Students are unsatisfied 

```python
# Create interactive widgets for filtering data
rating_threshold_slider = widgets.FloatSlider(value=7, min=0, max=10, step=0.1, description='Rating Threshold:')
```

```python
# Identify unsatisfied students 
unsatisfied_students = df[df['Average_Rating'] < 7]
print("\nUnsatisfied Students:\n", unsatisfied_students)
```

```python
# Create a bar chart for rating statistics
def update1_plots(rating_threshold):
    print("\nUnsatisfied Students:\n")
    plt.figure(figsize=(10, 6))
    plt.bar(df['Student Roll Num'], df['Average_Rating'], color='g', alpha=0.5, label='Average Rating')
    plt.bar(unsatisfied_students['Student Roll Num'], unsatisfied_students['Average_Rating'], color='r', alpha=0.7, label='Unsatisfied Students')
    plt.axhline(y=rating_threshold, color='r', linestyle='--', label=f'Rating Threshold ({rating_threshold})')
    plt.xlabel('Students')
    plt.ylabel('Average Rating')
    plt.title('Rating Statistics')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Create an interactive dashboard
interactive_dashboard = widgets.interactive(update1_plots, rating_threshold=rating_threshold_slider)

display(interactive_dashboard)
```

# Students are very satisfied

```python
# Identify very satisfied students 
very_satisfied_students = df[df['Average_Rating'] > 7]
print("very_satisfied_students", very_satisfied_students)
```

```python
# Create a bar chart for rating statistics
def update2_plots(rating_threshold):
    plt.figure(figsize=(10, 6))
    plt.bar(df['Student Roll Num'], df['Average_Rating'], color='g', alpha=0.5, label='Average Rating')
    plt.bar(very_satisfied_students['Student Roll Num'], very_satisfied_students['Average_Rating'], color='b', alpha=0.7, label='Very Satisfied Students')
    plt.axhline(y=rating_threshold, color='r', linestyle='--', label=f'Rating Threshold ({rating_threshold}))')
    plt.xlabel('Students')
    plt.ylabel('Average Rating')
    plt.title('Rating Statistics')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Create an interactive dashboard
interactive_dashboard = widgets.interactive(update2_plots, rating_threshold=rating_threshold_slider)

display(interactive_dashboard)
```

#  Compare Attended and Missed Students

```python
# Create interactive widgets for filtering data
attendance_threshold_slider = widgets.FloatSlider(value=7, min=0, max=10, step=0.1, description='Attendance Threshold:')
```

```python
# Function to update plots based on thresholds
def update_plots(attendance_threshold):
    plt.figure(figsize=(10, 6))
    plt.bar(df['Student Roll Num'], df['Attended_count'], label='Attended', color='b')
    plt.bar(df['Student Roll Num'], df['Missed_count'], label='Missed', color='r')
    plt.xlabel('Students')
    plt.ylabel('Sessions')
    plt.title('Attendance Statistics')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.axhline(y=attendance_threshold, color='r', linestyle='--', label=f'Attendance Threshold ({attendance_threshold})')
    plt.tight_layout()
    
    # Show the plots
    plt.show()

# Create an interactive dashboard
interactive_dashboard = widgets.interactive(update_plots, attendance_threshold=attendance_threshold_slider)

display(interactive_dashboard)
```
